```python
import matplotlib.pyplot as plt
 
 
labels = ['India', 'Canada', 'Japan', 'Australia', 'Russia']
sizes = [5, 19, 15, 14, 21]  # Add upto 100%

#https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.pie.html

# Plot the pie chart
# rotates the start of the pie chart by angle degrees counterclockwise from the x-axis.
#plt.pie(sizes, labels=labels, startangle=90)
# string format

plt.pie(sizes, labels=labels, autopct='%1.3f%%', startangle=90) #maintain two decimal places

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
 
# Display the graph onto the screen
plt.show()
```
