### Change the size of figures:
plt.rcParams["figure.figsize"] = (4, 8) will set graph figure with 400px in width and 800px in height.

```python
import matplotlib.pyplot as plt
 
# Changing default values for parameters individually
plt.rc('lines', linewidth=2, linestyle='-', marker='*')
plt.rc('figure', figsize= (5, 3))

#plt.rcParams["figure.figsize"] = (4, 8)
 
# Plot a line graph
plt.plot([10, 20, 30, 40, 50, 60, 70, 80])
# Add labels and title
plt.title("Interactive Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
 
```
